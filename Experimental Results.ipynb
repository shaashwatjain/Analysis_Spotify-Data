```python
#drop year coz its same for all
#drop instrumentalness coz mostly 0
#what are the features of the most likeable songs

df1 = df[df.popularity==1]
```

```python
#we use tsne to find similar objects. this reduces the dimension of the dataset to desired no of components.
#we had a lot of features like loudness,danceability,accousticness etc so we tried to reduce the dimension so we 
#could explain the trend better.
#we try to see different esults using PCA and tSNE
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

tsne = TSNE(n_components=2)
X= tsne.fit_transform(df1)
pca = PCA(n_components=2)
Y= pca.fit_transform(df1)
```

```python
#shape of the reduced features
print(X.shape)
print(Y.shape)
```

```python
#correlation between two features we got by using pca
import seaborn as sns
df2 = pd.DataFrame()
df2['firstcol'] = X[:,0]
df2['secondcol'] = X[:,1]
plt.figure(figsize=(8,8))

sns.scatterplot(x="firstcol", y="secondcol",data=df2)


```

```python
#correlation using two features we got using tsne.
df3 = pd.DataFrame()
df3['firstcol2'] = Y[:,0]
df3['secondcol2'] = Y[:,1]
plt.figure(figsize=(8,8))
sns.scatterplot(x="firstcol2", y="secondcol2",data=df3)
```

```python
#this is to see if we are getting some clusters based on the song features.
#dimensionality reduction is done using tsne. 
import hypertools as hyper
df= pd.read_csv("data.csv")
df = df.drop(labels = ["name","id","release_date","artists"],axis = 1)

df = df[df.year==2020]
df= df.reset_index()
df= df.drop(labels =['index'],axis = 1)
#df= df.drop(labels =['level_0'],axis = 1)
#df= df.drop(labels=['popularity'],axis = 1)
list1=[]
for i in range(0,len(df)):
    j= df['popularity'][i]
    if j>=0 and j<=20:
        list1.append(1)
    elif j>=21 and j<=40:
        list1.append(2)
    elif j>=41 and j<=60:
        list1.append(3)
    elif j>=61 and j<=80:
        list1.append(4)
    if j>=81 and j<=100:
        list1.append(5)
        
df['popularity_map']= list1
df= df.drop('popularity',axis =1)
target = df['popularity_map']
df2 = df.drop(labels= ['popularity_map','year','mode','duration_ms'], axis=1)
group = target.astype('str')
hyper.plot(df2,'.',group=group, reduce='TSNE',hue=group, ndims=2, legend=[1,2,3,4,5], title = "Clustering using t-SNE")
```
