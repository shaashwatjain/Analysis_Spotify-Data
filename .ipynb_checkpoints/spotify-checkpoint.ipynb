```python
# remove outliers which using z score with threshold 3

# some songs have explicit as 1 and should not be considered as outliers
int_df = int_df.drop(['explicit'], axis = 1) 
z = np.abs(stats.zscore(int_df))
# print(z)
threshold = 3
# print(np.shape(np.where(z > 3)))
new_df = int_df[(z < 3).all(axis=1)]
print("Dataframe of integer columns except explicit before removal of outliers", int_df.shape)
print("Dataframe of integer columns except explicit after removal of outliers", new_df.shape)
```

```python
plt.figure(figsize=(16, 8))
columns = ["acousticness","danceability","energy","speechiness","liveness","valence","instrumentalness"]
for col in columns:
    x = df.groupby("year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('Trends of Song features over the past century')
ax.set_ylabel('Values')
ax.set_xlabel('Year')
```

```python
# we plot trend of temp separately as it is not constrained to [0,1]

plt.figure(figsize=(10, 5))
sns.set(style="whitegrid")
columns = ["tempo"]
for col in columns:
    x = new_df.groupby("year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('Trend in tempo over the past century')
ax.set_ylabel('Tempo in bpm')
ax.set_xlabel('Year')

# we plot trend of temp separately as it is not constrained to [0,1]

plt.figure(figsize=(10, 5))
sns.set(style="whitegrid")
columns = ["loudness"]
for col in columns:
    x = new_df.groupby("year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('Trend in loudness over the past century')
ax.set_ylabel('Loudness in dB')
ax.set_xlabel('Year')
```
